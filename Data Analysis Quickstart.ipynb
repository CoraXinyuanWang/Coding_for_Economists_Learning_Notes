# Data Analysis Quickreview
In this chapter, we'll see analysis on a single dataframe using the Star Wars' characters dataset as an example. This chapter uses pandas and numpy packages. 

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# Set seed for random numbers
seed_for_prng = 78557
prng = np.random.default_rng(
    seed_for_prng
)
# prng is your random number generator object
```

```python
# Now, let's load data in
df = pd.read_csv("https://github.com/aeturrell/coding-for-economists/raw/main/data/starwars.csv", index_col=0,).dropna(subset =["species"])
# .dropna (), this removes any rows where the species column is missing (i.e., NaN)
df.info()
```

## Randomly selecting a sample using .sample 

```python
## .sample (n) randomly selects n rows, .sample (frac =0.4) selects 40% of the data, replace = True samples with replacement, and passing weights = selects a number of fraction with the probability given by the passed weight. (Note that the weights passed should have the same length as the dataframe)

df.sample (5)
```

## Summarize numerical values with .describe ()

```python
df.describe ()# It returns a table with the following statistics for each numeric column 
```

## Group variable values with .groupby ()

```python
df.groupby ("species")[["height","mass"]].mean () # df.groupby("grouping_column")[["column1", "column2"]].aggregation_function()
```

## Add transformed columns using .transform ()

Quite often, it's useful to put a column into a dataframe that is the result of an intermediate groupby and aggregation. Transform does this and returns a transformed column with the same shape as the original dataframe. Transform preserves the original index. 

```python
df ["mass_demean_species"] = df.groupby ("species")["mass"].transform (
    lambda x:x-x.mean()
)
df.head()
```

## Make quick charts with .plot.*()

```python
df.plot.scatter ("mass","height",alpha =0.5) # alpha =0.5 controls the transparency level 
```

```python
df.plot.box(column ="height") #column="height" tells Pandas to create a boxplot of the height column only.
```

```python
df ["height"].plot.kde(bw_method =0.3) # kde gives a smooth curve that estimates the probability density function of height values.

# bw_method controls the bandwidth — i.e., how smooth the curve is.
```
